```python
def is_same_page(page1,page2):
    try:
        system_messege = textwrap.dedent(""" Your job is to determine whether 2 pages are from same document or not
        , one of the ways to determine is by checking similar names, places, dates, etc. appear in both documents
        I they do return only boolean value "True" else "False"
        
        Another clear indicator is if the first few words in the page 2 are a 
        continuation of the alast words on page 1
        
        IMPORTANT : only return a boolean and nothing else ,dont give any introductory statement not even "content=" in the start
        """)
        llm = ChatOllama()
        p1 = aa_small['transcription_text'][0]
        p2 = aa_small['transcription_text'][1]
        
        e1 =aa_small['transcription_text'][3]
        e2 =aa_small['transcription_text'][4]

#         print("Page 1:",page1)
#         print("\n Page 2:",page2)

        mes = [SystemMessage(content =system_messege),
              HumanMessage(content=f'Page 1 :{p1} \n\n Page 2 :{p2}'),
              AIMessage(content ="True"),
              HumanMessage(content=f'Page 1 :{e1} \n\n Page 2 :{e1}'),
              AIMessage(content ="False"),
              HumanMessage(content=f'Page 1 :{page1} \n\n Page 2 :{page2}')
              ]

        res = llm.invoke(mes)
        return (str(res)=="content='True'")
    except Exceptin as e:
        print("Error",e)
        return False
```
